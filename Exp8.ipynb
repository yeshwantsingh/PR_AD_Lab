```python
import sklearn
from sklearn import datasets
from sklearn import tree
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
data, target = datasets.load_iris(as_frame=True, return_X_y=True)
```

```python
X, y = datasets.load_iris(return_X_y=True)
```

```python
train_x, test_x, train_y, test_y = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
decision_tree_classifier = tree.DecisionTreeClassifier(max_depth=3)
```

```python
decision_tree_classifier.fit(train_x, train_y)
```

```python
plt.style.use('classic')

fig = plt.figure(figsize=(15, 10))

ax = plt.gca()

tree.plot_tree(decision_tree_classifier, ax=ax)

plt.show()
```

```python
decision_tree_classifier.score(test_x, test_y)
```
