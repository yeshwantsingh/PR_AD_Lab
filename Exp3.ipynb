```python
import sklearn
from sklearn import datasets
from sklearn import linear_model
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
data, target = datasets.load_diabetes(as_frame=True, return_X_y=True)
```

```python
X, y = datasets.load_diabetes(return_X_y=True)
```

```python
data.head()
```

```python
plt.scatter(X[:,2], y)
plt.xlabel('BMI')
plt.ylabel('Diabetic Level')
plt.show()
```

```python
data_x = X[:, np.newaxis, 2]
data_y = y
```

```python
# train_x, test_x, train_y, test_y = data_x[:400], data_x[400:], data_y[:400], data_y[400:]
train_x, test_x, train_y, test_y = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
train_x.shape
```

```python
test_x.shape
```

```python
model_one_column = linear_model.LinearRegression(n_jobs=-1)
```

```python
model_one_column.fit(train_x, train_y)
```

```python
model_one_column.coef_, model_one_column.intercept_
```

```python
model_one_column.score(test_x, test_y)
```

```python
model_one_column.predict(np.array([[.2]]))
```

```python
model_all_column.fit(train_x, train_y)
```

```python
model_all_column.coef_, model_all_column.intercept_
```
